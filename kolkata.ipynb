```python
cleaned_dfs={'new_car_detail_df':'','new_car_overview_df':'','new_car_features_df':'','new_car_specs_df':''}
```

```python
def new_car_detail(csv_file,car_native):
                
    df1=pd.read_csv(csv_file)  
    it=[] 
    ft=[]
    bt=[]
    km=[]
    transmission=[]
    ownerNo=[]
    owner=[]
    oem=[]
    model=[]
    modelYear=[] 
    centralVariantId=[] 
    variantName=[]
    price=[]
    priceActual=[]
    priceSaving=[]
    priceFixedText=[] 
    trendingText_heading=[]
    trendingText_desc=[]
    
    for j in list(df1['new_car_detail']):
        #converting from json to dict
        res = ast.literal_eval(j)
        it.append(res['it'])
        ft.append(res['ft'])
        bt.append(res['bt'])
        km.append(res['km'])
        transmission.append(res['transmission'])
        ownerNo.append(res['ownerNo'])
        owner.append(res['owner'])
        oem.append(res['oem'])
        model.append(res['model'])
        modelYear.append(res['modelYear'])
        centralVariantId.append(res['centralVariantId'])
        variantName.append(res['variantName'])
        price.append(res['price'])
        priceActual.append(res['priceActual'])
        priceSaving.append(res['priceSaving'])
        priceFixedText.append(res['priceFixedText'])
        trendingText_heading.append(res['trendingText']['heading'])
        trendingText_desc.append(res['trendingText']['desc'])
    
    # Creating a new dataframe
    new_car_detail_df=pd.DataFrame({'id':[i for i in range(len(df1))],
                                    'car_from':[car_native for i in range(len(df1))],
                                    'it':it, 
                                    'ft':ft,
                                    'bt':bt,
                                    'km':km,
                                    'transmission':transmission,
                                    'ownerNo':ownerNo,
                                    'owner':owner,
                                    'oem':oem,
                                    'model':model,
                                    'modelYear':modelYear,
                                    'centralVariantId':centralVariantId,
                                    'variantName':variantName,
                                    'price':price,
                                    'priceActual':priceActual,
                                    'priceSaving':priceSaving,
                                    'priceFixedText':priceFixedText,
                                    'trendingText_heading':trendingText_heading,
                                    'trendingText_desc':trendingText_desc})

    cleaned_dfs['new_car_detail_df']=new_car_detail_df
    print('printing the shape of the new_car_detail_df dataset',new_car_detail_df.shape)
    print()
    print('Checking whether I have captured the information correctly...')
    print()
    s=int(input('Enter any record from the dataset you want to check'))
    print(f'The actual json string of the {s}th record:')
    print(df1['new_car_detail'].iloc[s])
    print()
    print(f'The modified {s}th record in the new_car_detail_df:')
    print(new_car_detail_df.iloc[s])
```

```python
spec_column=[]

for spec_json in list(df1['new_car_specs']):
    res = ast.literal_eval(spec_json)
    for f in range(len(res['top'])):
        spec_column.append(res['top'][f]['key'])
    p=0
    while p<len(res['data']):
        for i in res['data'][p]['list']:
            spec_column.append(i['key'])
        p+=1

spec_column.append('id')

# List containing only the unique spec names
# I will use this list as column names to create a new dataframe
spec_columns=sorted(list(set(spec_column)))
```

```python
len(spec_columns)
```

```python
spec_data=[]

for spec_json in list(df1['new_car_specs']):
    res = ast.literal_eval(spec_json)
    
    sp=[]
    for f in range(len(res['top'])):
        sp.append((res['top'][f]['key'],res['top'][f]['value']))
    p=0
    while p<len(res['data']):
        for i in res['data'][p]['list']:
            if i['key']!='Max Power':
                sp.append((i['key'],i['value']))
        p+=1
    #Arranging all elements in the list sp based on its 'key' i.e first element in the tuple
    spec_data.append(sorted(sp,key=lambda x: x[0]))
    #sp.clear()
spec_data[:5]
```
